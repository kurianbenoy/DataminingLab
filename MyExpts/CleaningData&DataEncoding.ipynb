```python
data=pd.DataFrame({'District':['Kottayam','Ernakulum','Idiki','Palakkad','Thiruvanathapuram','Karoor','Kollam', 'Test', 'Kasargod']})

encoder1=ce.OneHotEncoder(cols='District',handle_unknown='return_nan',return_df=True,use_cat_names=False)
data
```

```python
data_encoded1=pd.get_dummies(data=data,drop_first=True)
data_encoded1
```
